# Bayesian analysis Iris data

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.decomposition import PCA

# load the iris datasets
dataset = datasets.load_iris()
# fit a Gaussian Naive Bayes model to the data
model = GaussianNB()
model.fit(dataset.data, dataset.target)
# make predictions
expected = dataset.target
predicted = model.predict(dataset.data)
# summarize the fit of the model
print('         RAW DATA')
print ('iris.data.shape = ', iris.data.shape, ' = (samples,features)')
print ('\niris.data[0:5] \n',  iris.data[0:5])
print ('\niris.target_names\n',iris.target_names)
print ('\niris.target[0:10]\n', iris.target[0:10])
print('\n         METRICS CLASSIFICATION REPORT')
print('\n',metrics.classification_report(expected, predicted))
# Precision: the proportion of the predicted positive cases that were correct
# Recall (or also true positive rate): the proportion of positive cases that were correctly identified
# F1-Score: the harmonic mean of precision and recall
print('\n         CONFUSION MATRIX')
print('\n',metrics.confusion_matrix(expected, predicted))
# Principal Component Analysis 
dataset.data = iris.data
dataset.target = iris.target
target_names = iris.target_names
pca = PCA(n_components=2)
X_r = pca.fit(dataset.data).transform(dataset.data)
# Percentage of variance explained for each components
print('\nexplained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
# Plot PCA 
plt.figure()
for c, i, target_name in zip("rgb", [0, 1, 2], target_names):
    plt.scatter(X_r[dataset.target == i, 0], X_r[dataset.target == i, 1], c=c, label=target_name)
plt.legend()
plt.title('Principal Component Analysis of IRIS dataset')
plt.show()
```
